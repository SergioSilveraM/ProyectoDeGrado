# ***BenchMark***

```python
from reporte_metricas import ReporteMetricas

report = ReporteMetricas()
df = report.load()
```

## ***Validation***

### ***macro avg***

```python
dfval_macro = df.query("Type == 'val' and Class == 'macro avg'").drop(columns=['accuracy', 'log_loss', 'auc']).reset_index(drop=True)
dfval_macro.sort_values(by=['f1-score'], ascending=False, inplace=True)
dfval_macro
```

### ***weighted avg***

```python
dfval_weight = df.query("Type == 'val' and Class == 'weighted avg'").drop(columns=['accuracy', 'log_loss', 'auc']).reset_index(drop=True)
dfval_weight.sort_values(by=['f1-score'], ascending=False, inplace=True)
dfval_weight
```

### ***Global***

```python
df_val_global = df.query("Type == 'val' and Class == 'global'").drop(columns=['precision', 'recall', 'f1-score', 'support']).reset_index(drop=True)
df_val_global.sort_values(by=['auc'], ascending=False, inplace=True)
df_val_global
```

## ***Test***

### ***macro avg***

```python
dfval_macro = df.query("Type == 'test' and Class == 'macro avg'").drop(columns=['accuracy', 'log_loss', 'auc']).reset_index(drop=True)
dfval_macro.sort_values(by=['f1-score'], ascending=False, inplace=True)
dfval_macro
```

### ***weighted avg***

```python
dfval_weight = df.query("Type == 'test' and Class == 'weighted avg'").drop(columns=['accuracy', 'log_loss', 'auc']).reset_index(drop=True)
dfval_weight.sort_values(by=['f1-score'], ascending=False, inplace=True)
dfval_weight
```

During the ***benchmarking*** process, various classification models were compared by collecting relevant metrics on both the validation and test sets. The ***unbalanced XGBoost*** model showed the ***best overall performance*** in terms of ***F1-score*** on the validation set, reaching values of ***\$0.833\$ (macro)*** and ***\$0.892\$ (weighted)***, slightly outperforming the model balanced with ADASYN.

On the test set, although the balanced model showed a ***marginal improvement*** of ***\$+0.001\$ to \$+0.002\$*** in the ***F1-score***, these differences are not considered significant from a practical standpoint. Therefore, it is concluded that balancing ***did not substantially improve*** the model's performance and may even introduce bias through artificially generated minority samples via oversampling.

In addition to the aforementioned metrics, it was observed that although the balanced model offered a slight advantage in the 'No Payment' class (***\$+0.01\$***), ***it did not improve performance*** in the minority classes ('Partial Payment' and 'Full Payment'), where the ***F1-score*** was virtually equivalent. This reinforces the claim that balancing did not offer substantial benefits and, furthermore, due to the nature of the organization's operations, this marginal improvement in the majority class provides no practical advantage and unnecessarily increases model complexity.

Another key point is the ***stability of the XGBoost model*** across both sets, maintaining minimal differences between validation and test results, which indicates ***good generalization capacity***. From a practical perspective, synthetic oversampling was discarded as it did not provide relevant improvements, thereby avoiding the potential introduction of artificial noise that could affect the model's interpretability and robustness.

Regarding the comparison with other types of models, there is a ***systematic superiority*** of ***gradient-based*** models, such as ***XGBoost*** and ***LightGBM***, over classical approaches like ***decision trees***, ***logistic regression***, ***SVC***, and ***KNN***, and even over neural networks such as ***MLP***. This advantage is reflected both in average metrics (**macro** and **weighted**) and in consistency across validation and test sets. Moreover, this behavior aligns with recent literature, where *boosting* models have shown excellent capabilities in handling non-linear relationships, imbalanced datasets, and complex classification tasks in the financial and credit risk domains. In industrial settings, their use has also been consolidated due to their ***strong performance, speed, interpretability, and ease of fine-tuning***.

Finally, the model selected as the ***winner*** in the ***benchmarking*** process was the ***unbalanced XGBoost***, prioritizing both its technical performance and adaptability to production environments.
