# ***BenchMark***

```python
from reporte_metricas import ReporteMetricas

report = ReporteMetricas()
df = report.load()
```

## ***Validation***

### ***macro avg***

```python
dfval_macro = df.query("Type == 'val' and Class == 'macro avg'").drop(columns=['accuracy', 'log_loss', 'auc']).reset_index(drop=True)
dfval_macro.sort_values(by=['f1-score'], ascending=False, inplace=True)
dfval_macro
```

### ***weighted avg***

```python
dfval_weight = df.query("Type == 'val' and Class == 'weighted avg'").drop(columns=['accuracy', 'log_loss', 'auc']).reset_index(drop=True)
dfval_weight.sort_values(by=['f1-score'], ascending=False, inplace=True)
dfval_weight
```

### ***Global***

```python
df_val_global = df.query("Type == 'val' and Class == 'global'").drop(columns=['precision', 'recall', 'f1-score', 'support']).reset_index(drop=True)
df_val_global.sort_values(by=['auc'], ascending=False, inplace=True)
df_val_global
```

## ***Test***

### ***macro avg***

```python
dfval_macro = df.query("Type == 'test' and Class == 'macro avg'").drop(columns=['accuracy', 'log_loss', 'auc']).reset_index(drop=True)
dfval_macro.sort_values(by=['f1-score'], ascending=False, inplace=True)
dfval_macro
```

### ***weighted avg***

```python
dfval_weight = df.query("Type == 'test' and Class == 'weighted avg'").drop(columns=['accuracy', 'log_loss', 'auc']).reset_index(drop=True)
dfval_weight.sort_values(by=['f1-score'], ascending=False, inplace=True)
dfval_weight
```

Durante el proceso de ***benchmarking*** se compararon distintos modelos de clasificación, recopilando métricas relevantes tanto en los conjuntos de validación como de prueba. El modelo ***XGBoost sin balanceo*** presentó el ***mejor desempeño global*** en términos de ***F1-score*** en validación, alcanzando valores de ***$0.833$ (macro)*** y ***$0.892$ (weighted)***, superando ligeramente al modelo balanceado con ADASYN.

En el conjunto de prueba, si bien el modelo balanceado mostró una ***mejora marginal*** de ***$+0.001$ a $+0.002$*** en el ***F1-score***, estas diferencias no se consideran significativas desde un punto de vista práctico. Por tanto, se concluye que el balanceo ***no mejoró sustancialmente*** el rendimiento del modelo e incluso podría introducir sesgos a través de las muestras minoritarias generadas artificialmente mediante sobremuestreo.

Adicional a las métricas anteriores, se observó que, aunque el modelo balanceado ofreció una ligera ventaja en la clase 'No pago' (***$+0.01$***), ***no mejoró el rendimiento*** en las clases minoritarias ('Pago parcial' y 'Pago completo'), donde el ***F1-score*** fue prácticamente equivalente. Esto refuerza la afirmación de que el balanceo no ofreció beneficios sustanciales y, además, por la naturaleza de la actividad de la organización, esta mejora marginal en la clase mayoritaria no representa ningún beneficio, además de complejizar el modelo.

Otro punto a destacar es la ***estabilidad del modelo XGBoost*** a través de ambos conjuntos, manteniendo diferencias mínimas entre validación y test, lo que indica una ***buena capacidad de generalización***. Desde una perspectiva práctica, el uso de sobremuestreo sintético fue descartado al no aportar mejoras relevantes, evitando así la posible introducción de ruido artificial que afecte la interpretabilidad y robustez del modelo.

En cuanto a la comparación con otros tipos de modelos, se observa una ***superioridad sistemática*** de los modelos basados en ***gradiente***, como ***XGBoost*** y ***LightGBM***, frente a enfoques clásicos como ***árboles de decisión***, ***regresión logística***, ***SVC*** y ***KNN***, e incluso respecto a redes neuronales como el ***MLP***. Esta ventaja se refleja tanto en las métricas promedio (**macro** y **weighted**) como en la consistencia entre los conjuntos de validación y prueba. Además, este comportamiento es coherente con lo reportado en la literatura reciente, donde los modelos de *boosting* han demostrado una excelente capacidad para manejar relaciones no lineales, conjuntos de datos desbalanceados y tareas de clasificación complejas en el ámbito financiero y de riesgo crediticio. En el entorno industrial, su uso también se ha consolidado debido a su ***buen desempeño, velocidad, interpretabilidad y facilidad de ajuste fino***.

Finalmente, el modelo seleccionado como ***ganador*** en el proceso de ***benchmarking*** fue el ***XGBoost sin balanceo***, priorizando tanto su rendimiento técnico como su adaptabilidad a entornos productivos.
